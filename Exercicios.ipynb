```python
# Implementar uma pilha simples utilizando uma lista.
# 1. Crie uma pilha vazia (utilizando uma lista).
# 2. Empilhe (adicione) os n√∫meros 5, 10 e 15 na pilha.
# 3. Mostre o topo da pilha (√∫ltimo elemento).
# 4. Desempilhe (remova) o topo da pilha e mostre o novo topo.
# 5. Verifique se a pilha est√° vazia e imprima o resultado.

pilha = []

pilha.append(5)
pilha.append(10)
pilha.append(15)

print(pilha)

print("Topo da pilha",  pilha[-1])
print("Desempilhando o topo da pilha: ", pilha.pop())
print("Pilha atual: ", pilha)
print("A pilha est√° vazia? ", len(pilha) == 0)
```

```python
# Voc√™ foi contratado para desenvolver uma funcionalidade de ‚ÄúDesfazer/Refazer"
# para um editor de texto simples. O editor permite que o usu√°rio escreva palavras
# em um documento, e, caso cometa um erro, pode desfazer a √∫ltima a√ß√£o ou
# refaz√™-la caso mude de ideia.

print("--- Bloco de Notasüìù ---")

historico = [] 
refazer = []  

while True:
    print("\nTexto atual:", " ".join(historico) if historico else "[Vazio]")
    print("Op√ß√µes:")
    print("1 - Adicionar texto")
    print("2 - Refazer")
    print("3 - Sair")

    opcao = input("Escolha uma op√ß√£o: ")

    if opcao == "1":
        texto = input("Digite o texto: ")
        historico.append(texto)  
        refazer.clear()  
    elif opcao == "2":
        if historico:
            refazer.append(historico.pop())  
        else:
            print("Nada para desfazer!")
    elif opcao == "3":
        print("Fechando...")
        break
    else:
        print("Op√ß√£o inv√°lida! Tente novamente.")


```

```python
# Criar uma implementa√ß√£o simples de uma fila (Queue) usando listas.
# 1. Crie uma fila vazia.
# 2. Enfileire 3 n√∫meros inteiros na fila.
# 3. Mostre o primeiro elemento da fila.
# 4. Desenfileire o primeiro elemento e mostre o novo primeiro elemento.
# 5. Verifique se a fila est√° vazia ap√≥s as remo√ß√µes.

from collections import deque

fila = deque()

fila.append("A")
fila.append("B")
fila.append("C")
fila.append("D")
fila.append("E")

print(fila)
print("Primeiro Letra da fila", fila[0])
print("Removendo a primeira letra: ", fila.popleft())
print("Nova primeira letra da fila: ", fila[0])
print("A fila est√° vazia? ", len(fila) == 0)
```

```python
# Um consult√≥rio odontol√≥gico precisa gerenciar a ordem de atendimento dos
# pacientes. Para isso, foi decidido implementar uma fila onde os pacientes s√£o
# chamados na ordem em que chegaram (FIFO ‚Äì First In, First Out).

from collections import deque

fila = deque()

def exibir_fila():
    if fila:
        print("Fila de atendimento:")
        for i in range(len(fila)):
            print(f"{i + 1}. {fila[i]}")
    else:
        print("Nenhum paciente na fila.")

while True:
    print("--- Consult√≥rio Odontol√≥gico ü¶∑ ---")
    print("1 - Adicionar paciente")
    print("2 - Pr√≥ximo paciente")
    print("3 - Exibir fila")
    print("4 - Sair")
    
    opcao = input("Escolha uma op√ß√£o: ")

    match opcao:
        case "1":
            nome = input("Nome do paciente: ")
            fila.append(nome)
            print(f"{nome} foi adicionado √† fila.")
        
        case "2":
            if fila:
                chamado = fila.popleft()  # Remove o primeiro da fila
                print(f"Chamando {chamado} para o atendimento.")
            else:
                print("Nenhum paciente na fila.")

        case "3":
            exibir_fila()

        case "4":
            print("Fechando...")
            break

        case _:
            print("Inv√°lido. Tente Novamente")

```

```python
# Uma impressora trabalha com um sistema de fila para processar documentos
# enviados para impress√£o. Cada documento enviado entra no final da fila e ser√°
# impresso na ordem em que chegou.
# ‚Ä¢ Crie uma fila vazia para representar a fila de impress√£o.
# ‚Ä¢ Enfileire tr√™s documentos (nomes de arquivos) na fila.
# ‚Ä¢ Mostre qual documento ser√° impresso primeiro.
# ‚Ä¢ Desenfileire o primeiro documento e mostre o pr√≥ximo da fila.
# ‚Ä¢ Verifique se ainda h√° documentos na fila ap√≥s as remo√ß√µes.

from collections import deque

fila = deque()

def exibir_fila():
    if fila:
        print("Fila de atendimento:")
        for i in range(len(fila)):
            print(f"{i + 1}. {fila[i]}")
    else:
        print("Nenhum paciente na fila.")

while True:
    print("--- Impressoraüñ®Ô∏è ---")
    print("1 - Adicionar um Documento")
    print("2 - Pr√≥ximo Documento")
    print("3 - Exibir fila de documentos √† serem impressos")
    print("4 - Sair")
    
    opcao = input("Escolha uma op√ß√£o: ")

    match opcao:
        case "1":
            nome = input("Nome do Arquivo: ")
            fila.append(nome)
            print(f"{nome} foi adicionado √† fila.")
        
        case "2":
            if fila:
                chamado = fila.popleft() 
                print(f"Pr√≥ximo documento para impress√£o: {chamado}.")
            else:
                print("Nenhum  na Documento na fila.")

        case "3":
            exibir_fila()

        case "4":
            print("Fechando...")
            break

        case _:
            print("Inv√°lido. Tente Novamente")

```

```python
# Uma central de atendimento telef√¥nico gerencia chamadas em espera usando
# uma pilha. Isso significa que a √∫ltima chamada recebida ser√° a primeira a ser
# atendida.
# ‚Ä¢ Crie uma pilha vazia para armazenar as chamadas em espera.
# ‚Ä¢ Empilhe tr√™s chamadas com identificadores num√©ricos.
# ‚Ä¢ Mostre qual chamada ser√° atendida primeiro.
# ‚Ä¢ Atenda a √∫ltima chamada recebida e mostre a nova chamada no topo da
# pilha.
# ‚Ä¢ Verifique se ainda h√° chamadas na pilha ap√≥s os atendimentos.

pilha = []

pilha.append(101)  
pilha.append(102)  
pilha.append(103)  

print(f"Chamada a ser atendida primeiro: {pilha[-1]}")

atendida = pilha.pop()
print(f"Atendendo chamada: {atendida}")

if pilha:
    print(f"Nova chamada no topo: {pilha[-1]}")
else:
    print("N√£o h√° mais chamadas na pilha.")

if pilha:
    print("Ainda h√° chamadas em espera.")
else:
    print("Todas as chamadas foram atendidas.")
```
